```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from structure import Structure
import json
```

```python
payload = json.dumps({
    "ScenarioData": {
    "type": "Dispersion",
    "frequency": 496,
},
"Layers": [
    {
            "type": "Ambient Incident Layer",
            "permittivity": 11
    },
    {
        "type": "Semi Infinite Anisotropic Layer",
        "material": "Quartz",
        "rotationX": 0,
        "rotationY": 90,
        "rotationZ": 0,
    }
],
})
payload = json.loads(payload)
```

```python
structure = Structure()
structure.get_scenario(payload.get('ScenarioData'))
structure.get_layers(payload.get("Layers", None))
quartz_layer = structure.layers[1]
```

```python
fig, axs = plt.subplots(
      1,2, figsize = (15,15), subplot_kw= dict(projection = 'polar')
  )
im1 = axs[0].pcolormesh(tf.math.real(structure.azimuthal_angle), tf.math.real(structure.incident_angle), tf.math.imag(quartz_layer.profile['propagation'][:,:,0])**2., cmap='magma')
im2 = axs[1].pcolormesh(tf.math.real(structure.azimuthal_angle), tf.math.real(structure.incident_angle), tf.math.imag(quartz_layer.profile['propagation'][:,:,1])**2., cmap='magma')
cbar1 = plt.colorbar(im1, ax=axs[0], shrink=0.4)
cbar2 = plt.colorbar(im2, ax=axs[1], shrink=0.4)
axs[0].grid(linewidth=0.1)
axs[1].grid(linewidth=0.1)
plt.show()
```
